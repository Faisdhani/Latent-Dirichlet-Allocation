```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from gensim import corpora
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel
import pyLDAvis.gensim_models as gensimvis
import matplotlib.pyplot as plt
import re
```

```python
# Download resource NLTK
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
# Load data dari CSV
data = pd.read_csv('/content/scrapped_data.csv')  # Pastikan file data.csv sudah diunggah ke Colab
data.head()
```

Case Folding (Lowercase)

```python
# Case Folding (lowercase)
data['lowercase'] = data['content'].str.lower()
data.head()
```

Tokenizing

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt') # Download resource 'punkt' for tokenization

def word_tokenize_wrapper(text):
  return word_tokenize(text)

data['tokens'] = data['lowercase'].apply(word_tokenize_wrapper)
data.head()
```

Normalization (Mengganti Kata Slang)

```python
# Normalization (contoh untuk bahasa Indonesia)
slang_dict = {'gak': 'tidak', 'ga': 'tidak', 'tdk': 'tidak', 'yg': 'yang'}  # contoh

def normalize(tokens, slang_dict):
    return [slang_dict[word] if word in slang_dict else word for word in tokens]

data['tokens_normalized'] = data['tokens'].apply(lambda x: normalize(x, slang_dict))
data.head()
```

Filtering (Stopword Removal)

```python
# Filtering (Stopword Removal)
stop_words = set(stopwords.words('indonesian'))  # Anda bisa menambahkan stopword tambahan di sini

data['tokens_filtered'] = data['tokens_normalized'].apply(
    lambda x: [word for word in x if word not in stop_words and word.isalpha()]
)
data.head()
```

Stemming

```python
# Stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()

data['tokens_stemmed'] = data['tokens_filtered'].apply(
    lambda x: [stemmer.stem(word) for word in x]
)
data.head()
```

```python
data.to_csv('hasil_preprocessing_satusehat.csv', index=False)
```

Menghitung Frekuensi Kata TF-IDF

```python
# Import library yang diperlukan
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from gensim import corpora
from wordcloud import WordCloud
from sklearn.decomposition import PCA

# Membaca file CSV yang sudah diproses
df_processed = pd.read_csv('hasil_preprocessing_satusehat.csv')

# Pastikan kolom yang digunakan adalah 'tokens_stemmed' (dikonversi ke string)
processed_data = df_processed['tokens_stemmed'].apply(lambda x: ' '.join(eval(x)))

# Inisialisasi TfidfVectorizer dengan max_features=1000
vectorizer = TfidfVectorizer(max_features=1000)

# Terapkan TF-IDF ke data yang sudah diproses
tfidf_matrix = vectorizer.fit_transform(processed_data)

# Ambil istilah/fitur TF-IDF
tfidf_terms = vectorizer.get_feature_names_out()

# Mengonversi matrix TF-IDF ke format Gensim corpus (list of tuples)
corpus = [[(idx, tfidf_value) for idx, tfidf_value in zip(doc.indices, doc.data)] for doc in tfidf_matrix]

# Simpan corpus dan terms untuk digunakan di bagian LDA
corpus_file = 'tfidf_corpus.mm'
dictionary_file = 'tfidf_dictionary.dict'

# Simpan corpus dan dictionary menggunakan Gensim
corpora.MmCorpus.serialize(corpus_file, corpus)
dictionary = corpora.Dictionary([tfidf_terms])
dictionary.save(dictionary_file)

print("TF-IDF telah dihitung dan disimpan.")

# Visualisasi Skor TF-IDF Tertinggi
# Ambil skor TF-IDF dan kata-katanya
tfidf_scores = np.array(tfidf_matrix.sum(axis=0)).flatten()
tfidf_words = tfidf_terms

# Urutkan kata berdasarkan skor TF-IDF
sorted_indices = np.argsort(tfidf_scores)[::-1]
top_n = 20  # Jumlah kata teratas yang ingin ditampilkan

# Pilih top_n kata dan skor TF-IDF-nya
top_words = tfidf_words[sorted_indices][:top_n]
top_scores = tfidf_scores[sorted_indices][:top_n]

# Visualisasikan dengan bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_words, top_scores, color='skyblue')
plt.xlabel('Skor TF-IDF')
plt.title(f'Top {top_n} Kata dengan Skor TF-IDF Tertinggi')
plt.gca().invert_yaxis()  # Membalik urutan agar kata teratas berada di atas
plt.show()

# Visualisasi WordCloud
# Buat dictionary kata dan skor TF-IDF
tfidf_scores_dict = {term: tfidf_matrix.getcol(idx).sum() for idx, term in enumerate(tfidf_terms)}

# Generate WordCloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(tfidf_scores_dict)

# Tampilkan WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("WordCloud Berdasarkan Skor TF-IDF")
plt.show()

# Visualisasi PCA
# Reduksi dimensi menggunakan PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(tfidf_matrix.toarray())

# Visualisasikan hasil PCA
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.6, c='blue', edgecolors='k')
plt.title('PCA dari TF-IDF Vektor')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

Pemodelan LDA (Latent Dirichlet Allocation)

```python
# Membuat dictionary dan corpus
dictionary = corpora.Dictionary(data['tokens_stemmed'])
corpus = [dictionary.doc2bow(text) for text in data['tokens_stemmed']]

# LDA Model
num_topics = 8  # Jumlah kluster
lda_model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=10)

# Tampilkan topik yang ditemukan beserta kata-kata dan skor
print("Topik yang Ditemukan:")
for idx, topic in lda_model.show_topics(num_topics=num_topics, formatted=False):
    # Tampilkan kata-kata dan skor per kata dalam topik
    topic_words = [f"{word}: {prob:.4f}" for word, prob in topic]
    print(f"Topik {idx}: {', '.join(topic_words)}")
```

```python
# Menampilkan kata-kata teratas dari setiap topik
num_topics = 8  # Misalkan ada 8 topik (sesuaikan dengan jumlah topik Anda)

for topic_id in range(num_topics):
    # Ambil 10 kata teratas di topik ini
    top_words = lda_model.show_topic(topic_id, topn=5)

    # Tampilkan kata-kata dan bobotnya
    print(f"Topik {topic_id + 1}:")
    for word, weight in top_words:
        print(f"{word}: {weight:.4f}")
    print("\n" + "-"*40 + "\n")
```

Evaluasi Model Menggunakan Coherence Score

```python
# Evaluasi Model
coherence_model_lda = CoherenceModel(
    model=lda_model, texts=data['tokens_stemmed'], dictionary=dictionary, coherence='c_v'
)
coherence_score = coherence_model_lda.get_coherence()
print(f"Coherence Score: {coherence_score}")
```

Evaluasi Jumlah Topik yang Optimal

```python
# Plotting coherence score untuk jumlah topik yang berbeda (opsional)
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=1):
    coherence_values = []
    for num_topics in range(start, limit, step):
        model = LdaModel(corpus=corpus, num_topics=num_topics, id2word=dictionary, passes=10)
        coherence_model = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherence_model.get_coherence())
    return coherence_values

limit = 20
coherence_values = compute_coherence_values(dictionary, corpus, data['tokens_stemmed'], limit)

# Plotting
plt.plot(range(2, limit), coherence_values)
plt.xlabel("Jumlah Topik")
plt.ylabel("Coherence Score")
plt.title("Evaluasi Jumlah Topik")
plt.show()
```
